```python
#task1
class Calculator:
    def __init__(self,n1,op,n2):
        print("Let’s Calculate!")
        self.num1 = int(n1)
        self.operator = op
        self.num2 = int(n2)
        print("Value 1:",self.num1)
        print("Operator:",self.operator)
        print("Value 2:",self.num2)
    def add(self):
        result = self.num1+self.num2
        print("Result:",result)
    def substract(self):
        result = self.num1-self.num2
        print("Result:",result)
    def multiply(self):
        result = self.num1*self.num2
        print("Result:",result)
    def divide(self):
        result = self.num1/self.num2
        print("Result:",result)
        
inp=input("Enter the operation: ").split()
task=Calculator(inp[0],inp[1],inp[-1])
if task.operator=="+":
    task.add()
elif task.operator=="-":
    task.substract()
elif task.operator=="*":
    task.multiply()
elif task.operator=="/":
    task.divide()
```

```python
#2
class Customer:
    def __init__(self,name):
        self.name = name
        self.purchased_items=[]
    def greet(self,name = None):
        if name != None:
            print(f"Hello {name}!")
        else:
            print("Hello!")
    def purchase(self,*items):
        for i in items:
            self.purchased_items.append(i)
        print(self.name,", you purchased",len(self.purchased_items),"item(s)")
        for x in self.purchased_items:
            print(x)
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#5

class Student:
    def __init__(self,name="default student"):
        self.name=name
    def quizcalc(self,*mark):
        result=0
        for i in mark:
            result+=i
        score = result/3
        self.score=score
    def printdetail(self):
        print(f"Hello {self.name}")
        print(f"Your average quiz score is {self.score}")
s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail()
```

```python
#9
class Patient:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def add_Symptom(self,*args):
        ans=""
        disease=""
        for i in args:
            ans+= i+", "
        for x in range(len(ans)-2):
            disease+=ans[x]
        self.symptom = disease
    def printPatientDetail(self):
        print(f"Name: {self.name}\nAge: {self.age}\nSymptoms: {self.symptom}")
p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
#13
class Batsman:
    def __init__(self,*args):
        lst=[]
        for i in args:
            lst.append(str(i))
        if len(lst)==3:
            self.name = lst[0]
            self.run =int(lst[1])
            self.ball = int(lst[2])
        elif len(lst)==2:
            if lst[0].isalpha()==False:
                self.name="New Batsman"
                self.run=int(lst[0])
                self.ball=int(lst[1])
            elif len(lst)==1:
                self.name=lst[0]
                self.run=6101
                self.ball=7380
    def setName(self,name):
        self.name=name
    def battingStrikeRate(self):
        sr = (self.run/self.ball)*100
        self.strike_rate=sr
        return self.strike_rate
    def printCareerStatistics(self):
        print(f"Name: {self.name}\nRuns Scored: {self.run}, Balls Faced:{self.ball}")
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#14
class EPL_Team:
    def __init__(self,club_name,song="No Slogan",title_no=0):
        self.club = club_name
        self.song = song
        self.total_title = title_no
    def changeSong(self,song):
        self.song = song
    def showClubInfo(self):
        name = self.club
        song = self.song
        no_title = self.total_title
        ans= "Name: {0}\nSong: {1}\nTotal No of title: {2}".format(name,song,no_title)
        return ans
    def increaseTitle(self,title_no=None):
        if title_no==None:
            self.total_title += 1
        elif title_no!=None:
            self.total_title+=int(title_no)
manu = EPL_Team('Manchester United', 'Glory Glory Man United')
chelsea = EPL_Team('Chelsea')
print('===================')
print(manu.showClubInfo())
print('##################')
manu.increaseTitle()
print(manu.showClubInfo())
print('===================')
print(chelsea.showClubInfo())
chelsea.changeSong('Keep the blue flag flying high')
print(chelsea.showClubInfo())
```

```python
#16
class Author:
    def __init__(self,*args):
        if len(args)==0:
            self.name = "Default"
        elif len(args)>0:
            if len(args)==1:
                self.name = args[0]
            elif len(args)>1:
                self.name = args[0]
                self.books = args[1:len(args)]
    def addBooks(self,*books):
        self.books= books
    def printDetails(self):
        print("Authors Name: ",self.name)
        print("--------")
        print("List of books:")
        for book in self.books:
            print(book)
    def changeName(self,name):
        self.name = name
auth1 = Author('Humayun Ahmed')
auth1.addBooks('Deyal', 'Megher Opor Bari')
auth1.printDetails()
print('===================')
auth2 = Author()
print(auth2.name)
auth2.changeName('Mario Puzo')
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian')
print('===================')
auth2.printDetails()
print('===================')
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain')
auth3.printDetails()
```
