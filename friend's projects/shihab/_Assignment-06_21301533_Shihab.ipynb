```python
#1
class Student:
    ID=0
    def __init__(self,name,department,age,cgpa):
        Student.ID+=1
        self.name=name
        self.department=department
        self.age=age
        self.cgpa=cgpa
    def get_details(self):
        print("ID:",Student.ID)
        print("Name:", self.name)
        print('Department:',self.department)
        print('Age:', self.age)
        print('CGPA:', self.cgpa)
    @classmethod
    def from_String(cls,string):
        name, department, age, cgpa = string.split('-')
        obj=cls(name,department,age,cgpa)
        return obj
            
    
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

print("Class variables are declared within a class, but outside a constructor. Instance variables are created when an object is created")
print('Instance method does whats prvided in the parameter and is bound to the object of a class. Class method is called on the class itself and not on a specific object.it can only access class variable.')

```

```python
#2
class Assassin:
    counter=0
    def __init__(self,name,rate):
        Assassin.counter+=1
        self.name=name
        self.rate=rate
    @classmethod
    def failureRate(cls,name,f_rate):
        s_rate=100-f_rate
        return cls(name,s_rate)
    
    @classmethod
    def failurePercentage(cls,name,fp_rate):
        sp_rate=100-fp_rate
        return cls(name,sp_rate)
    
    def printDetails(self):
        print("Name:",self.name)
        print('Success rate:',str(self.rate)+"%")
        print('Total number of Assassin:',Assassin.counter)

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#5
from datetime import date

class Employee:
    def __init__(self,name,workingPeriod):
        self.name=name
        self.workingPeriod=workingPeriod
    @classmethod
    def employeeByJoiningYear(cls,name,year):
        workingPeriod=date.today().year-year
        return cls(name,workingPeriod )
    @staticmethod
    def experienceCheck(year,gender):
        if year>3:
            if gender=="male":
                return("He is experienced")
            else:
                return("She is experienced")
        else:
            if gender=="male":
                return("He is not experienced")
            else:
                return("She is not experienced")
        
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#7
class Cat:
    Number_of_cats=0
    def __init__(self,color,action):
        self.color=color
        self.action=action
        Cat.Number_of_cats+=1
    def changeColor(self,color):
        self.color=color
    def printCat(self):
        print(self.color,"cat is",self.action)
    @classmethod
    def no_parameter(cls):
        color="White"
        action="sitting"
        return cls(color,action)
    @classmethod
    def first_parameter(cls,color):
        action="sitting"
        return cls(color,action)
    @classmethod
    def second_parameter(cls,action):
        color="Grey"
        return cls(color,action)
    
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#8
class Cylinder:
    height = 18
    radius = 5
    def __init__(self, radius, height):
        default = "Default radius={} and height={}."
        print(default.format(Cylinder.radius, Cylinder.height))
        Cylinder.height = height
        Cylinder.radius = radius
        updated = "Updated: radius={} and height={}."
        print(updated.format(radius, height))
  
    @classmethod
    def swap(cls,height,radius):
        return cls(radius,height)
  
    @classmethod
    def changeFormat(cls,str):
        values = str.split("-")
        return cls(float(values[0]), float(values[1]))

    @staticmethod
    def area(radius,height):
        PI = 3.14
        res = 2 * PI * radius * (radius + height)
        print("Area: " + str(res))
  
    @staticmethod
    def volume(radius, height):
        PI = 3.14
        res =  PI * (radius * radius) * height
        print("Volume: " + str(res))
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
